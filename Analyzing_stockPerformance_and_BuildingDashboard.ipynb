```python
import yfinance as yf

# Fetch Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="1y")  
tesla_data.head()
```

## Tesla Stock Data

The code above fetches Tesla's stock data for the past year using the `yfinance` library. The output shows a preview of the data, including open, high, low, close prices, and volume.


```python
import requests
from bs4 import BeautifulSoup

url = "https://finance.yahoo.com/quote/TSLA/financials?p=TSLA"

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')


try:
   
    revenue_elements = soup.find_all("div", class_="D(ib) Mend(20px)")
    
    for element in revenue_elements:
      
        text = element.get_text(strip=True)
        if "Revenue" in text:
            revenue = text
            break
    else:
        revenue = "Revenue data not found"

except Exception as e:
    revenue = f"An error occurred: {e}"

print("Tesla Revenue Data:", revenue)
```

## GameStop Stock Data

The code above fetches GameStop's stock data for the past year using the `yfinance` library. The output shows a preview of the data, including open, high, low, close prices, and volume.


```python
import dash
from dash import dcc, html
import plotly.graph_objs as go

app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Tesla Stock and Revenue Dashboard"),
    dcc.Graph(
        id='tesla-stock-graph',
        figure={
            'data': [
                go.Scatter(
                    x=tesla_data.index,
                    y=tesla_data['Close'],
                    mode='lines',
                    name='Tesla Stock Price'
                )
            ],
            'layout': {
                'title': 'Tesla Stock Price'
            }
        }
    ),
])

if __name__ == '__main__':
    app.run_server(debug=True)
```

## Tesla Stock and Revenue Dashboard

The dashboard visualizes Tesla's stock price and revenue data. Please note that the full functionality of the dashboard may require running the app separately.


```python
import dash
from dash import dcc, html
import plotly.graph_objs as go
import yfinance as yf

gamestop = yf.Ticker("GME")
gamestop_data = gamestop.history(period="1y")
app = dash.Dash(__name__)
app.layout = html.Div([
    html.H1("GameStop Stock Price Dashboard"),
    
    dcc.Graph(
        id='gamestop-stock-graph',
        figure={
            'data': [
                go.Scatter(
                    x=gamestop_data.index,
                    y=gamestop_data['Close'],
                    mode='lines',
                    name='GameStop Stock Price'
                )
            ],
            'layout': {
                'title': 'GameStop Stock Price',
                'xaxis': {'title': 'Date'},
                'yaxis': {'title': 'Price (USD)'}
            }
        }
    )
])

if __name__ == '__main__':
    app.run_server(debug=True)
```
